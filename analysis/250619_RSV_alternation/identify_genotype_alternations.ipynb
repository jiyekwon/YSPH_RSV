```python
import numpy as np
import pandas as pd
import glob
import os
```

```python
# Read all TSV files from the /data/ folder

rsvatab = pd.read_csv('data/rsv-a_metadata_2025-06-19T1926.tsv', sep='\t')
rsvbtab = pd.read_csv('data/rsv-b_metadata_2025-06-19T1927.tsv', sep='\t')
rsvatab['genotype'] = 'RSVA'
rsvbtab['genotype'] = 'RSVB'

# Concatenate all case tables
cases_df = pd.concat([rsvatab, rsvbtab], ignore_index=True)

# Ensure date column is datetime
cases_df['date'] = pd.to_datetime(cases_df['sampleCollectionDate'],yearfirst=True,format="mixed")

# Extract year-month for grouping
cases_df['year_month'] = cases_df['date'].dt.to_period('M')
cases_df['year'] = cases_df['date'].dt.to_period('Y')
```

```python
# Summarise counts per month for each location
country_year = cases_df.groupby(['geoLocCountry','year']).size().reset_index(name='case_count').pivot(index='geoLocCountry', columns='year', values='case_count')

```

```python
summary = cases_df.groupby(['geoLocCountry', 'year_month','genotype']).size().reset_index(name='case_count')

print(summary)
```
