```python
import pandas as pd
```

```python
metafile = "data/rsv_metadata.txt"
samplefile = "data/RSVA.samples"
outfile = "data/dates.txt"
metacol="date"
```

```python
samples = open(samplefile).read().split()
samples
```

```python
meta = pd.read_csv(metafile,sep="\t")
meta['date'] = pd.to_datetime(meta['date'],format="%Y-%m-%d")
```

```python
meta = meta[meta['name'].isin(samples)]
```

```python
mindate = min(meta['date'])
meta['epidays'] = pd.to_timedelta(meta['date']-mindate).dt.days
meta[['name','epidays']]
```

```python
meta[['name','epidays']].to_csv(outfile,sep="\t",index=False,header=False)
```
