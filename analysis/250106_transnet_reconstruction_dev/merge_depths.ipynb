```python
import pandas as pd
```

```python
target = "RSVA"
samplefile = "data/RSVA.samples"
```

```python
samples = open(samplefile).read().split()
```

```python
alldepth = None
si=-1
for sample in samples:
    si = si+1
    sdepth = pd.read_csv("data/{}_{}_depth.txt.gz".format(sample,target),sep="\t").iloc[:,2]
    if alldepth is None:
        alldepth = pd.DataFrame(index=range(1,len(sdepth)),columns=samples)
    alldepth[sample] = sdepth
```

```python
alldepth.to_csv("{}_alldepth.txt".format(target),sep="\t")
```
