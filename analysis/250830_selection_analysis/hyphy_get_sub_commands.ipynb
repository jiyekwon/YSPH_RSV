```python
import gffutils
import pyfaidx
from Bio.Seq import Seq
import subprocess
from Bio import SeqIO

from Bio import Phylo

import pandas as pd
import json
```

```python

targets = ["RSVB", "RSVA"]


with open('hyphy_sub_all_clade_analyses.sh', 'w') as f:
    for target in targets:

        gff = f'refs/{target}.gff3'
        
        db = gffutils.create_db(gff, dbfn=':memory:')
        geneids = [cds.attributes['gene'][0] for cds in db.features_of_type('CDS', order_by='start')]

        nextcladeF = f"{target}_nextclade.tsv"
        df_clades = pd.read_csv(nextcladeF, usecols=['seqName', 'clade', 'qc.overallStatus'], sep="\t")
        df_clades['majorclade'] = df_clades['clade'].str.split('.').str[:4].str.join('.')
        
        clades = df_clades['clade'].unique()
        majorclades = df_clades['majorclade'].unique()

        for gene in geneids:
            fasta = f"refs/{target}_aligned_{gene}.fasta"
            
            for clade in clades:
                #print(f"Running {target} {gene} {clade}")
                runstring = " ".join(['python hyphy_run_clade_selection_analyses.py ',
                            f'-a geneseqs/{target}_{gene}_codonaln_PGCOE.fasta ',
                            f'-C {nextcladeF} -c {clade} ',
                            f'-p geneseqs/{target}_{gene}_tree_pruned.nwk ',
                            f'-g {gene} -t {target} ',
                            '-o hyphy_relax/hyphy'])
                print(runstring, file=f)
            for clade in majorclades:
                runstring = " ".join(['python hyphy_run_clade_selection_analyses.py ',
                            f'-a geneseqs/{target}_{gene}_codonaln_PGCOE.fasta ',
                            f'-C {nextcladeF} -c {clade} -M',
                            f'-p geneseqs/{target}_{gene}_tree_pruned.nwk ',
                            f'-g {gene} -t {target} ',
                            '-o hyphy_relax/hyphy_M'])
                print(runstring, file=f)
```
