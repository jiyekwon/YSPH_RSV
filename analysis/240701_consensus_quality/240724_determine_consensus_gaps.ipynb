```python
import numpy as np
import pandas as pd
import pysam 
```

```python
#define bad regions:
maxbad = 0.2

consqualf = "RSVB-filtered-aligned-qual.txt"

consqual = pd.read_table(consqualf)
consqual['badprop']=consqual['bad'] / consqual[['good','ambiguous','bad']].sum(axis=1)
badpos = consqual['pos'][consqual['badprop'] > maxbad]

fi = None
li = None
badwins = list()
for i in badpos:
    if fi is None:
        fi = i
    elif i > li+1:
        if li-fi > 20:
            badwins.append((fi,li,li-fi))
        fi = i
    li = i
if li-fi > 20:
    badwins.append((fi,li,li-fi))
badwins = pd.DataFrame(badwins, columns =['start', 'end', 'size'])
```

```python
badwins.to_csv("RSVB_consensus_gaps.txt",sep="\t",index=False)
```
