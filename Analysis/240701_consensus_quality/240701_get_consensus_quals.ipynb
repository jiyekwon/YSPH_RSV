```python
import allel 
import numpy as np

import pandas as pd
from Bio import SeqIO

from plotnine import *
```

```python
ref="RSVA"
fasta = "./{}-filtered-aligned.fasta".format(ref)
outfile = "./{}-filtered-aligned-qual".format(ref)
seqs = list(SeqIO.parse(fasta, "fasta"))
```

```python
reflen = len(seqs[0].seq)
nseqs = len(seqs)

```

```python
good = [0] * reflen
ambig = [0] * reflen
bad = [0] * reflen
for seq in seqs:
    for i in range(0,len(seq.seq)):
        b = seq.seq[i]
        if   b in ['A','C','T','G']: good[i] = good[i]+1
        elif b == 'N': bad[i] = bad[i]+1    
        else: ambig[i] = ambig[i]+1    
consquals = pd.DataFrame({"pos":list(range(1,reflen+1)),"good":good,"ambiguous":ambig,"bad":bad})
```

```python
consquals.to_csv(outfile+".txt",index=False,header=True,sep="\t")
```

```python
consqualsm = consquals.melt(id_vars="pos")

qplot = ggplot(consqualsm, aes(x="pos",y="value",fill="variable")) + \
            geom_bar(stat="identity",color=None) + \
            scale_fill_manual(values={"good":"green",
                                      "ambiguous":"orange",
                                      "bad":"red"},
                             name=None) + \
            ggtitle(ref+" consensus calls") +\
            theme(legend_position="bottom",legend_title=element_blank(),axis_title=element_blank())
qplot

qplot.save((outfile+".png"),width=300,height=100,dpi=400,units="mm")
```
